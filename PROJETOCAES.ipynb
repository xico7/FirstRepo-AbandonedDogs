```python

df['Ageint'] = df.Age

df['Ageint'] = df['Ageint'].fillna(0)

df.Ageint = df.Ageint.astype(str)

df['Ageint'].replace('mes','m',regex=True,inplace=True)
df['Ageint'].replace('ano','a_',regex=True,inplace=True)
df['Ageint'].replace('e','',regex=True,inplace=True)
df['Ageint'].replace('s','',regex=True,inplace=True)
df['Ageint'] = df['Ageint'].apply(str).str.replace('\(|\)','')



```

Age is written with full words so i have to replace and separate the years and months in order to get the dogs age in days

```python
df1 = pd.DataFrame(df.Ageint.str.split('_',1).tolist(),
                                   columns = ['Year','Month'])
df1['Month'].fillna(value = 0,inplace=True)
df1.Month.replace('','0',regex=True,inplace=True)
df1.Month = df1.Month.astype(str)
df1.Month.replace('m','',regex=True,inplace=True)
df1.Month = df1.Month.astype(int)


df1['daysinmonth'] = 30
values = df1.Month * df1.daysinmonth
df1['Monthvalues'] = values

# Calculates the 2nd part of the dogs age and shows the result in number of days 
# if dog has only 1 variable of age i.e. 2 years exactly or less than 1 year this code does nothing
```

```python
df2 = pd.DataFrame(df1.Year.str.split(' ',1).tolist(),
                                   columns = ['Number','MorY'])

df2.MorY.replace('a','365',regex=True,inplace=True)
df2.MorY.replace('m','30',regex=True,inplace=True)
df2['MorY'].fillna(value = 0,inplace=True)
df2.Number = df2.Number.astype(int)
df2.MorY = df2.MorY.astype(int)

values2 = df2.Number * df2.MorY
df2['MonthandYearvalues'] = values2

# Calculates the 1st part of the dogs age wether its years or months and shows the result in number of days 
```

```python
df['ValueofAgeInt'] = df2.MonthandYearvalues + df1.Monthvalues
```

```python
df3 = df[['ValueofAgeInt', 'Adoptiondate']].copy()
df3.ValueofAgeInt.dropna(how='any', inplace=True)
df3.dropna(inplace = True)
df3 = df3[df3.ValueofAgeInt != 0]
```
